# Sub-Goal 1: Assess Customer Engagement

## Question: What is the average number of purchases made by customers?
## Metric: Calculate the mean number of purchases per customer.

```python
import pandas as pd

# Load the dataset
dataset_path = r'C:\Users\B.RENUKA\Desktop\Winter Sem 2024\Project\Metrics\online_retail_customer_churn.csv'
df = pd.read_csv(dataset_path)

# Calculate the total number of purchases
total_purchases = df['Num_of_Purchases'].sum()
print(f"Total Number of Purchases: ",total_purchases)

# Calculate the number of unique customers
num_customers = df['Customer_ID'].nunique()
print(f"No of customers : ",num_customers)

# Calculate the average number of purchases per customer
average_purchases_per_customer = total_purchases / num_customers

print("Average number of purchases per customer:", average_purchases_per_customer)
```

## Question: How does the average transaction amount vary across different customer segments?
## Metric: Compare the average transaction amount for different age groups, genders, and income levels.


```python
import pandas as pd

# Load the dataset
dataset_path = r'C:\Users\B.RENUKA\Desktop\Winter Sem 2024\Project\Metrics\online_retail_customer_churn.csv'
df = pd.read_csv(dataset_path)

# Calculate overall average transaction amount for age
avg_transaction_amount_age = df['Average_Transaction_Amount'].mean()

# Calculate overall average transaction amount for gender
avg_transaction_amount_gender = df.groupby('Gender')['Average_Transaction_Amount'].mean()

# Calculate overall average transaction amount for income level
avg_transaction_amount_income = df['Average_Transaction_Amount'].mean()

print("Average Transaction Amount (Age Group):", avg_transaction_amount_age)
print("\nAverage Transaction Amount (Gender):")
print(avg_transaction_amount_gender)
print("\nAverage Transaction Amount (Income Level):", avg_transaction_amount_income)
```

## Question: How does the number of support contacts impact customer satisfaction?
## Metric: Analyze the correlation between the number of support contacts and satisfaction scores.


```python
import pandas as pd
import numpy as np

# Load the dataset
dataset_path = r'C:\Users\B.RENUKA\Desktop\Winter Sem 2024\Project\Metrics\online_retail_customer_churn.csv'
df = pd.read_csv(dataset_path)

# Step 1: Compute the means of both variables
mean_support_contacts = df['Num_of_Support_Contacts'].mean()
mean_satisfaction_score = df['Satisfaction_Score'].mean()

# Step 2: Compute the differences from the mean
diff_support_contacts = df['Num_of_Support_Contacts'] - mean_support_contacts
diff_satisfaction_score = df['Satisfaction_Score'] - mean_satisfaction_score

# Step 3: Compute the product of the differences
product_diff = diff_support_contacts * diff_satisfaction_score

# Step 4: Compute the sum of the products
sum_product_diff = product_diff.sum()

# Step 5: Compute the squared differences from the mean
squared_diff_support_contacts = diff_support_contacts ** 2
squared_diff_satisfaction_score = diff_satisfaction_score ** 2

# Step 6: Compute the sum of squared differences
sum_squared_diff_support_contacts = squared_diff_support_contacts.sum()
sum_squared_diff_satisfaction_score = squared_diff_satisfaction_score.sum()

# Step 7: Compute the square root of the sum of squared differences for each variable
sqrt_sum_squared_diff_support_contacts = np.sqrt(sum_squared_diff_support_contacts)
sqrt_sum_squared_diff_satisfaction_score = np.sqrt(sum_squared_diff_satisfaction_score)

# Step 8: Compute the Pearson correlation coefficient
pearson_correlation_coefficient = sum_product_diff / (sqrt_sum_squared_diff_support_contacts * sqrt_sum_squared_diff_satisfaction_score)

print("Pearson correlation coefficient between Number of Support Contacts and Satisfaction Scores:", pearson_correlation_coefficient)
```
